## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Check value counts of Special_Considerations field
application_df['SPECIAL_CONSIDERATIONS'].value_counts()
```

```python
# Check value counts of STATUS field
application_df['STATUS'].value_counts()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#Additionally remove STATUS and SPECIAL CONSIDERATIONS due to Value Counts providing most of data is only 1 category and seems like it may not be relevant.
application_df = application_df.drop(columns = ['EIN','NAME','STATUS','SPECIAL_CONSIDERATIONS'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = application_df['APPLICATION_TYPE'].value_counts()
app_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = app_counts[app_counts < 700].index

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts[classification_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = classification_counts[classification_counts <300].index

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.nunique()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_dummies_df = pd.get_dummies(application_df, dtype = int)
application_dummies_df.head()
```

```python
# Reviewinig list of columns to identify columns to remove (for categorical need 1 less than number of categories)
application_dummies_df.columns
```

```python
# Drop addditional columns to remove categorical redundancies
application_dummies_df = application_dummies_df.drop(columns = ['APPLICATION_TYPE_Other','AFFILIATION_Other','CLASSIFICATION_Other','USE_CASE_Other','ORGANIZATION_Trust',
                                                                'INCOME_AMT_1-9999'])
```

```python
# Split our preprocessed data into our features and target arrays
y = application_dummies_df['IS_SUCCESSFUL']
X = application_dummies_df.drop (columns = 'IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Reviewing feature data columns and checking number of model inputs (#columns)
X.head()
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer - #pre-processed dataset has 34 inputs/columns
nn.add(tf.keras.layers.Dense(units=30, activation="leaky_relu", input_dim=len(X_train_scaled[0])))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=20, activation="leaky_relu"))

## Third hidden layer
nn.add(tf.keras.layers.Dense(units=10, activation="leaky_relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
